```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("C:/Users/Poojitha/Desktop/flood.csv")
df
```

```python
# 1. Salary Spread by Department
plt.figure(figsize=(8,5))
df.boxplot(column="DamsQuality", by="Landslides", grid=False, showfliers=False)
plt.title("DamsQuality Spread by Landslides")
plt.suptitle("") # remove default title
plt.xlabel("Landslides")
plt.ylabel("DamsQuality")
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Landslides Distribution
plt.figure(figsize=(8,5))
plt.hist(df['Landslides'], bins=20, color="orange", edgecolor="black")
plt.title("Landslides Distribution")
plt.xlabel("Landslides")
plt.ylabel("Frequency")
plt.show()
```

```python
# 3. Monsoon Distribution
plt.figure(figsize=(8,5))
plt.hist(df['MonsoonIntensity'], bins=7, color="skyblue", edgecolor="black")
plt.title("Monsoon Probability Distribution")
plt.xlabel("Monsoon Intensity")
plt.ylabel("Frequency")
plt.show()
```

```python
# 4. Watersheds vs InadequatePlanning
plt.figure(figsize=(8,5))
plt.hexbin(df['Watersheds'], df['InadequatePlanning'], gridsize=25, cmap="Greens", alpha=0.8)
plt.colorbar(label="Counts")
plt.title("Watersheds vs InadequatePlanning (Hexbin Heatmap)")
plt.xlabel("Watersheds")
plt.ylabel("InadequatePlanning")
plt.show()
```

```python
# 5. AgriculturalPractices by PopulationScore
population_count = df["PopulationScore"].value_counts()
plt.figure(figsize=(8,5))
plt.barh(population_count.index, population_count.values, color="navy")
plt.title("AgriculturalPractices by PopulationScore")
plt.xlabel("AgriculturalPractices")
plt.ylabel("PopulationScore")
plt.show()
```

```python
# 6. Average WetlandLoss by PoliticalFactors
avg_bonus = df.groupby("WetlandLoss")["PoliticalFactors"].mean()
plt.figure(figsize=(8,5))
plt.plot(avg_bonus.index, avg_bonus.values, marker="o", linestyle="--", color="red")
plt.title("Average WetlandLoss by Department")
plt.xlabel("PoliticalFactors")
plt.ylabel("WetlandLoss")
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. Average DamsQuality by Siltation
avg_damsquality = df.groupby("DamsQuality")["Siltation"].mean().sort_values()
plt.figure(figsize=(8,5))
avg_damsquality.plot(kind="barh", color="purple", edgecolor="black")
plt.title("Average DamsQuality by Siltation")
plt.xlabel("Average Salary")
plt.ylabel("DamsQuality")
plt.show()
```

```python
# 8. DrainageSystems Count by FloodProbability 
drainagesys_count = df["Deforestation"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(drainagesys_count, labels=drainagesys_count.index, autopct="%1.1f%%", startangle=90)
plt.title("DrainageSystems Count by FloodProbability")
plt.show()
```

```python
# 9. Plot distribution of FloodProbability
plt.figure(figsize=(8,5))
sns.histplot(df['FloodProbability'], bins=20, kde=True, color='c', edgecolor='black')
plt.title('Flood Probability Distribution')
plt.xlabel('FloodProbability')
plt.ylabel('Count')
plt.show()
```

```python
from pyspark.sql.functions import avg, min, max

# Average flood probability
df_clean.agg(avg("FloodProbability")).show()

# Minimum and maximum for each feature
stat_cols = ['MonsoonIntensity', 'RiverManagement', 'Deforestation', 'Urbanization', 'FloodProbability']
for colname in stat_cols:
    df_clean.agg(min(colname).alias(f"min_{colname}"), max(colname).alias(f"max_{colname}")).show()
```

```python
for feature in [c for c in df_clean.columns if c != "FloodProbability"]:
    corr = df_clean.stat.corr(feature, "FloodProbability")
    print(f"Correlation between {feature} and FloodProbability: {corr}")
```

```python
import os
import shutil
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Create Spark session
spark = SparkSession.builder \
    .appName("FloodDataProject") \
    .config("spark.hadoop.io.nativeio.enabled", "false") \
    .getOrCreate()

# Load the dataset (point directly to CSV file)
input_path = "C:/Users/Poojitha/Desktop/flood.csv"
df = spark.read.csv(input_path, header=True, inferSchema=True)

# Create DataFrame of locations with high flood probability (> 0.7)
output_dir = "C:/Users/Poojitha/Desktop/BDA/Flood_HighProb_temp"
final_file = "C:/Users/Poojitha/Desktop/BDA/Flood_HighProb.csv"

high_prob_df = df.filter(col("FloodProbability") > 0.7)

try:
    # Write Spark output to temp folder
    high_prob_df.coalesce(1) \
        .write.option("header", True) \
        .mode("overwrite") \
        .csv(output_dir)
    
    # Find the part file Spark created
    for file in os.listdir(output_dir):
        if file.startswith("part-") and file.endswith(".csv"):
            part_file = os.path.join(output_dir, file)
            # Move and rename to desired file
            shutil.move(part_file, final_file)
            break

    # Clean up temp directory
    shutil.rmtree(output_dir)

    print(f"✅ Saved flood cases with probability > 0.7 into: {final_file}")

except Exception as e:
    print("⚠️ Spark CSV write failed, using Pandas fallback...", e)
    high_prob_df.toPandas().to_csv(final_file, index=False)
    print(f"✅ Saved flood cases with probability > 0.7 using Pandas fallback.")
```
